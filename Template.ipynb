# Array

## Two Pointers

```python
def twoSum(nums: list[int], target: int) -> list[int]:
    l, r = 0, len(nums) - 1

    while l < r:
        cur = nums[l] + nums[r]
        if cur == target:
            return [l, r]
        elif cur < target:
            l += 1
        else:
            r -= 1
    return []

# Example usage
nums = [2,7,11,15]
target = 9
print(twoSum(nums, target))
```

## Sliding Window


```python
def lengthOfLongestSubstring(s: str) -> int:
    l = 0
    res = 0
    visited = {}

    for r in range(len(s)):
        # shrink the window if the character is already visited and is within the current window
        if s[r] in visited and visited[s[r]] >= l: 
            l = visited[s[r]] + 1

        res = max(res, r - l + 1)
        visited[s[r]] = r

    return res

# Example usage
s = "abcabcbb"
print(lengthOfLongestSubstring(s))
```

# Prefix Sum


```python
class Solution(object):
    def subarraySum(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        prefix = 0
        count = 0
        hashMap = {0:1}

        for num in nums:
            prefix += num
            if prefix - k in hashMap:
                # If prefix - k is in the hashMap, it means there are hashMap[prefix - k] subarrays that sum to k
                count += hashMap[prefix - k]
            hashMap[prefix] = hashMap.get(prefix, 0) + 1

        return count
        
# Example usage
nums = [1, 1, 1]
k = 2
print(Solution().subarraySum(nums, k))
```

# Linked List

## Fast and Slow Pointers


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None
class Solution(object):
    def detectCycle(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head or not head.next:
            return None

        slow, fast = head, head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                fast = head # Move fast to the beginning of the list

                # Move both pointers at the same speed until they meet; the point they meet at will be the start of the cycle
                while slow != fast: 
                    slow = slow.next
                    fast = fast.next
                return slow

        return None
```

# Reverse a linked list


```python
class Solution(object):
    def reverseList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        if not head or not head.next:
            return head

        prev, cur = None, head

        while cur:
            temp = cur.next
            cur.next = prev
            prev = cur
            cur = temp

        return prev
```

## Dummy Node Template


```python
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        dummy = ListNode(0,head)
        prev, cur = dummy, head
        while cur:
            if cur.val == val:
                cur = cur.next
                prev.next = cur
            else:
                prev = prev.next
                cur = cur.next
                
        return dummy.next
```

# Binary Tree

## DFS (recursive)

```python
def dfs(root):
    if not root:
        return
    
    ans = 0

    # do logic
    dfs(root.left)
    dfs(root.right)
    return ans
```

## DFS (iterative)

```python
def dfs(root):
    stack = [root]
    ans = 0

    while stack:
        node = stack.pop()
        # do logic
        
        # Preorder: add right child first so that left child is processed first
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)

    return ans
```
