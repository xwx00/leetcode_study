## Linked List

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

There are two kinds of **linked list**: **singly linked list** and **doubly linked list**. In a singly linked list, each node contains a value and a reference to the next node. In a doubly linked list, each node contains a value, a reference to the next node, and a reference to the previous node.

Linked lists are very similar to arrays, but to find the i-th element in a linked list, we need to traverse the list from the head node to the i-th node, which takes O(i) time. In contrast, we can access the i-th element in an array in O(1) time. 

However, linked lists are more flexible than arrays because they can easily grow and shrink in size without the need for resizing or copying elements. It's also easier to insert and delete elements in a linked list than in an array, especially when the list is large whereas in an array, we may need to shift elements to maintain the order after insertion or deletion.


```python
class ListNode():
    def __init__(self,val):
        self.val = val
        self.next = None

class MyLinkedList():
    def __init__(self):
        self.size = 0
        self.head = None
    
    def createLinkedList(self, nums):
        if not nums:
            return None

        dummy = ListNode(0)
        cur = dummy 
        for num in nums:
            node = ListNode(num)
            cur.next = node
            cur = cur.next
            self.size += 1

        self.head = dummy.next

    def printLinkedList(self, index):
        res = []
        cur = self.head
        if index > self.size:
            index = self.size
            
        for _ in range(index):
            res.append(cur.val)
            cur = cur.next
        return res
         
    def get(self, index: int):
        if index >= self.size:
            return -1

        cur = self.head
        for _ in range(index):
            cur = cur.next
        
        return cur.val

    def addAtHead(self, val: int):
        new_head = ListNode(val)
        new_head.next = self.head
        self.head = new_head
        self.size += 1

    def addAtTail(self, val: int):
        cur = self.head
        while cur.next:
            cur = cur.next

        cur.next = ListNode(val)
        self.size += 1

    def addAtIndex(self,index: int ,val: int):
        if index == 0:
            self.addAtHead(val)

        elif index == self.size:
            self.addAtTail(val)

        elif index > self.size:
            return

        else:
            prev, cur = ListNode(0), self.head
            prev.next = cur
            for _ in range(index):
                prev = cur
                cur = cur.next

            node = ListNode(val)
            node.next = cur
            prev.next = node
            self.size += 1

    def deleteAtIndex(self, index: int):
        if index >= self.size:
            return -1
        elif index == 0:
            self.head = self.head.next
            self.size -=  1
        else:
            cur = self.head
            for _ in range(index-1):
                cur = cur.next 

            cur.next = cur.next.next
            self.size -= 1
            

```

```python
myLinkedList = MyLinkedList()
myLinkedList.createLinkedList([1,2,3,4,5])
print(myLinkedList.printLinkedList(2))

for i in range(6):
    print(myLinkedList.get(i))

myLinkedList.addAtHead(9)
print(myLinkedList.printLinkedList(9))

myLinkedList.addAtTail(9)
print(myLinkedList.printLinkedList(9))

myLinkedList.addAtIndex(2,3)
print(myLinkedList.printLinkedList(9))

myLinkedList.deleteAtIndex(5)
print(myLinkedList.printLinkedList(9))
```

## Linked List Cyle


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        if not head or not head.next:
            return False
        
        slow, fast = head, head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
                
        return False
```

```python
def has_cycle(head):
    # Initialize slow & fast pointers
    slow = head
    fast = head

    # Loop condition (avoid None errors)
    while slow and fast and fast.next:
        slow = slow.next          # move slow pointer one step
        fast = fast.next.next     # move fast pointer two steps

        # Condition to check (depends on problem)
        if slow == fast:
            return True

    return False   # adjust return based on problem
```

## Dummy Node Template

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        dummy = ListNode(0,head)
        prev, cur = dummy, head
        while cur:
            if cur.val == val:
                cur = cur.next
                prev.next = cur
            else:
                prev = prev.next
                cur = cur.next
                
        return dummy.next
```

## Doubly Linked List Template

```python
class DoublyListNode:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None
```

```python
class DoublyListNode:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None

## NOTICE: Every function should have return statement, even if it is None. Otherwise, it will cause error when we call the function in the future.
class MyLinkedList:

    def __init__(self):
        self.size = 0
        self.head = None
        

    def get(self, index: int) -> int:
        if index > self.size - 1:
            return -1
        
        cur = self.head
        for _ in range(index):
            cur = cur.next
        return cur.val
        

    def addAtHead(self, val: int) -> None:
        if not self.head:
            self.head = DoublyListNode(val)
            self.size += 1
            return
        
        new_head = DoublyListNode(val)
        new_head.next = self.head
        self.head.prev = new_head
        self.head = new_head
        self.size += 1
        return

    def addAtTail(self, val: int) -> None:
        if not self.head:
            self.head = DoublyListNode(val)
            self.size += 1
            return
        
        tail = self.head
        for _ in range(self.size-1):
            tail = tail.next

        node = DoublyListNode(val)
        tail.next = node
        node.prev = tail
        self.size += 1
        return

    def addAtIndex(self, index: int, val: int) -> None:
        if index > self.size:
            return -1
        elif index == 0:
            self.addAtHead(val)
        elif index == self.size:
            self.addAtTail(val)
        else:
            node = DoublyListNode(val)
            cur = self.head
            for _ in range(index):
                cur = cur.next

            prev = cur.prev
            node.next = cur
            node.prev = cur.prev
            cur.prev = node
            prev.next = node
            self.size += 1
            return

    def deleteAtIndex(self, index: int) -> None:
        if index > self.size - 1:
            return -1
        
        dummy = DoublyListNode(0)
        dummy.next = self.head
        self.head.prev = dummy

        prev, cur = dummy, self.head
        for _ in range(index):
            prev = prev.next
            cur = cur.next

        prev.next = cur.next

        if cur.next:
            cur.next.prev = prev

        self.size -= 1
        self.head = dummy.next
        return
        


# Your MyLinkedList object will be instantiated and called as such:
# obj = MyLinkedList()
# param_1 = obj.get(index)
# obj.addAtHead(val)
# obj.addAtTail(val)
# obj.addAtIndex(index,val)
# obj.deleteAtIndex(index)
```
