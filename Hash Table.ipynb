# Introduction

There are two different kinds of hash tables: hash set and hash map.

- The **hash set** is one of the implementations of a set data structure to store no repeated values.
- The **hash map** is one of the implementations of a map data structure to store (key, value) pairs.

By choosing a proper hash function, the hash table can achieve wonderful performance in both insertion and search.

# Design Hash Table
Use a hash function to map keys to buckets.

1. Use hash function to decide which bucket to put the key in when we want to insert a key.

2. When we want to search for a key, the hash table will use the same hash function to find the corresponding bucket and search only in the specific bucket.

# Aggregate by Key

````python
def aggregate_by_key_hashmap(keys):
    hashmap = {}

    for key in keys:
        if key in hashmap:
            # update existing value
            hashmap[key] += 1
        else:
            # initialize value
            hashmap[key] = 1

    return hashmap
````

# Example usage


``` python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        char_map = {}   # character -> last seen index
        l = 0           # left pointer of window
        res = 0         # result

        for r in range(len(s)):
            char = s[r]

            # 1. Handle duplicate (shrink window)
            if char in char_map and char_map[char] >= l:
                l = char_map[char] + 1

            # 2. Update hashmap
            char_map[char] = r

            # 3. Update result
            res = max(res, r - l + 1)

        return res
```
