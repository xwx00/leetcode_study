## Definition

A heap is a binary tree that meets the following criteria:

- Is a complete binary tree;
- The value of each node must be no greater than (for Min-Heap) or no less than (for Max-Heap) the value of its child nodes.

````python
class MinHeap:
    def __init__(self):
        self.heap = []

    def size(self):
        return len(self.heap)

    def peek(self):
        return self.heap[0] if self.heap else None

    def push(self, val):
        self.heap.append(val)
        self._heapify_up(len(self.heap) - 1)

    def pop(self):
        if not self.heap:
            return None
        
        # swap root with last element
        self.heap[0], self.heap[-1] = self.heap[-1], self.heap[0]
        min_val = self.heap.pop()
        
        self._heapify_down(0)
        return min_val

    # ===== helper functions =====
    def _heapify_up(self, i):
        while i > 0:
            parent = (i - 1) // 2
            if self.heap[i] < self.heap[parent]:
                self.heap[i], self.heap[parent] = self.heap[parent], self.heap[i]
                i = parent
            else:
                break

    def _heapify_down(self, i):
        n = len(self.heap)
        while True:
            left = 2 * i + 1
            right = 2 * i + 2
            smallest = i

            if left < n and self.heap[left] < self.heap[smallest]:
                smallest = left
            if right < n and self.heap[right] < self.heap[smallest]:
                smallest = right

            if smallest == i:
                break

            self.heap[i], self.heap[smallest] = self.heap[smallest], self.heap[i]
            i = smallest

    def __str__(self):
        return str(self.heap)
````

## Operations

| Operation            | Python Code                     | Time Complexity | Space Complexity | Notes |
|---------------------|--------------------------------|------------------|------------------|------|
| Create empty heap   | `heap = []`                    | O(1)             | O(1)             | Just an empty list |
| Heapify (build)     | `heapq.heapify(arr)`           | O(n)             | O(1)             | In-place build |
| Push (insert)       | `heapq.heappush(heap, x)`      | O(log n)         | O(1)             | Bubble up |
| Pop (remove min)    | `heapq.heappop(heap)`          | O(log n)         | O(1)             | Bubble down |
| Peek (get min)      | `heap[0]`                      | O(1)             | O(1)             | No removal |
| Push + Pop          | `heapq.heappushpop(heap, x)`   | O(log n)         | O(1)             | More efficient than push + pop |
| Replace (pop+push)  | `heapq.heapreplace(heap, x)`   | O(log n)         | O(1)             | Always pops first |
| n largest elements  | `heapq.nlargest(k, arr)`       | O(n log k)       | O(k)             | Uses min heap of size k |
| n smallest elements | `heapq.nsmallest(k, arr)`      | O(n log k)       | O(k)             | Uses max heap of size k |

````python
import heapq

# Create an array
minHeap = []

# Heapify the array into a Min Heap
heapq.heapify(minHeap)
print(minHeap)  # Output: []

# Add 3, 1, 2 to the Min Heap
heapq.heappush(minHeap, 3)
heapq.heappush(minHeap, 1)
heapq.heappush(minHeap, 2)
print("minHeap:",minHeap)  # Output: [1, 3, 2]

# Get the top element of the Min Heap
print("minHeap[0]:",minHeap[0])  # Output: 1

# Delete the top element of the Min Heap
popNum = heapq.heappop(minHeap)
print("popNum:",popNum)  # Output: 1

# Check the top element of the Min Heap after deletion
print("Top element after deletion:",minHeap[0])  # Output: 2

# Check the size of the Min Heap
size = len(minHeap)
print("Size of minHeap:",size)  # Output: 2
````

# Heap Sort Template

````python
def heap_sort(self, lst: List[int]) -> None:
        """
        Mutates elements in lst by utilizing the heap data structure
        """
        def max_heapify(size,i):
            l, r = 2*i + 1, 2*i + 2
            largest = i
            if l < size and lst[l] > lst[largest]:
                largest = l
            if r < size and lst[r] > lst[largest]:
                largest = r
            if largest != i:
                lst[largest], lst[i] = lst[i], lst[largest]
                max_heapify(size,largest)

        for i in range(len(lst) // 2 - 1, -1, -1):
            max_heapify(len(lst), i)

        for i in range(len(lst) - 1, -1, -1):
            lst[i], lst[0] = lst[0], lst[i]
            max_heapify(i,0) # note that we decrease the size by 1 every time
````

````python
from typing import List
class Solution:
    def heap_sort(self, lst: List[int]) -> None:
        """
        Mutates elements in lst by utilizing the heap data structure
        """
        def max_heapify(size,i):
            l, r = 2*i + 1, 2*i + 2
            largest = i
            if l < size and lst[l] > lst[largest]:
                largest = l
            if r < size and lst[r] > lst[largest]:
                largest = r
            if largest != i:
                lst[largest], lst[i] = lst[i], lst[largest]
                max_heapify(size,largest)

        for i in range(len(lst) // 2 - 1, -1, -1):
            max_heapify(len(lst), i)

        for i in range(len(lst) - 1, -1, -1):
            lst[i], lst[0] = lst[0], lst[i]
            max_heapify(i,0) # note that we decrease the size by 1 every time

def test_basic():
    sol = Solution()

    lst = [4, 10, 3, 5, 1]
    sol.heap_sort(lst)
    assert lst == [1, 3, 4, 5, 10]

    lst = [1]
    sol.heap_sort(lst)
    assert lst == [1]

    lst = []
    sol.heap_sort(lst)
    assert lst == []

def test_sorted_and_reverse():
    sol = Solution()

    lst = [1, 2, 3, 4, 5]
    sol.heap_sort(lst)
    assert lst == [1, 2, 3, 4, 5]

    lst = [5, 4, 3, 2, 1]
    sol.heap_sort(lst)
    assert lst == [1, 2, 3, 4, 5]

import random

def test_random():
    sol = Solution()

    for _ in range(10):
        lst = [random.randint(-1000, 1000) for _ in range(100)]
        expected = sorted(lst)

        sol.heap_sort(lst)

        assert lst == expected

def test_negative_numbers():
    sol = Solution()

    lst = [-1, -3, -2, 0, 2, 1]
    sol.heap_sort(lst)
    assert lst == [-3, -2, -1, 0, 1, 2]

def test_duplicates():
    sol = Solution()

    lst = [2, 3, 2, 1, 4, 1]
    sol.heap_sort(lst)
    assert lst == [1, 1, 2, 2, 3, 4]

def test_edge_cases():
    sol = Solution()

    lst = [2, 1]
    sol.heap_sort(lst)
    assert lst == [1, 2]

    lst = [1, 1, 1, 1]
    sol.heap_sort(lst)
    assert lst == [1, 1, 1, 1]

def run_all_tests():
    test_basic()
    test_sorted_and_reverse()
    test_duplicates()
    test_negative_numbers()
    test_random()
    test_edge_cases()
    print("All tests passed!")

run_all_tests()
````

```python
class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        if not intervals:
            return 0
        
        intervals.sort(key = lambda x: x[0])
        heap = []
        heapq.heappush(heap,intervals[0][1])
        
        for i in range(1,len(intervals)):
            start, end = intervals[i]
            
            if start >= heap[0]:
                heapq.heappop(heap)
                
            heapq.heappush(heap,end)
            
        return len(heap)
```

## Furthest Building You Can Reach


可以。这题的标准做法就是 贪心 + 最小堆。

核心想法

从 heights[i] 到 heights[i+1]：

- 如果下降或持平，不需要资源
- 如果上升 diff = heights[i+1] - heights[i]，就需要：
- 要么用砖
    - 要么用梯子
    - 贪心关键

梯子应该优先留给 最大的上升高度
砖头优先去填 较小的上升高度

因为：

    梯子不管多高都只花 1 个
    砖头花费正比于高度差

怎么实现这个贪心？

我们一路遍历时，先把每次上升都“假装用梯子”：

- 把 diff 放进一个 最小堆
- 如果堆的大小超过 ladders，说明梯子不够用了
- 那就从堆里弹出最小的 diff，改成用砖头支付

这样就保证了：

- 堆里留下的始终是“最大的那些 climb” → 用梯子
- 被弹出去的小 climb → 用砖头

```python 
class Solution:
    def furthestBuilding(self, heights: List[int], bricks: int, ladders: int) -> int:
        min_heap = []

        for i in range(len(heights) - 1):
            diff = heights[i + 1] - heights[i]

            if diff <= 0:
                continue

            heapq.heappush(min_heap, diff)

            if len(min_heap) > ladders:
                bricks -= heapq.heappop(min_heap)

            if bricks < 0:
                return i

        return len(heights) - 1
```

##  Find Median from Data Stream


标准解：两个堆

思路：

small：大顶堆，存较小的一半
large：小顶堆，存较大的一半

Python 没有大顶堆，所以 small 里存相反数。
```python
import heapq

class MedianFinder:

    def __init__(self):
        self.small = []  # max heap via negatives
        self.large = []  # min heap

    def addNum(self, num: int) -> None:
        heapq.heappush(self.small, -num)
        
        if self.small and self.large and (-self.small[0] > self.large[0]):
            val = -heapq.heappop(self.small)
            heapq.heappush(self.large, val)
        
        if len(self.small) > len(self.large) + 1:
            val = -heapq.heappop(self.small)
            heapq.heappush(self.large, val)
        
        if len(self.large) > len(self.small):
            val = heapq.heappop(self.large)
            heapq.heappush(self.small, -val)

    def findMedian(self) -> float:
        if len(self.small) > len(self.large):
            return -self.small[0]
        return (-self.small[0] + self.large[0]) / 2
```
