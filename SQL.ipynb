# Introduction


---

## 1️⃣ Database Software

**Definition:**  
The system that manages databases.

**Examples:**
- :contentReference[oaicite:0]{index=0}  
- :contentReference[oaicite:1]{index=1}  
- :contentReference[oaicite:2]{index=2}  

👉 Handles storage, querying, security, and performance.

---

## 2️⃣ Database

**Definition:**  
A container that holds related data.

**Purpose:**
- Separate data by business domain or use case

**Example:**
- `us_stock_db`
- `london_stock_db`

👉 Each database is independent.

---

## 3️⃣ Schema

**Definition:**  
A logical grouping of tables and database objects.

**Key Points:**
- Organizes tables by logic (like a blueprint)
- Can include metadata and structure definitions
- Helps manage permissions and structure

⚠️ **Note:**
- In MySQL → *Schema = Database* (they are the same)
- In PostgreSQL / SQL Server → Schema is a separate layer

---

## 4️⃣ Table

**Definition:**  
The core structure that stores data.

**Contains:**
- Columns (name, type, constraints)
- Rows (actual records)

**Example Tables:**
- `users`
- `products`
- `orders`

👉 Defines how data is structured.

---

## 5️⃣ Data

**Definition:**  
The actual stored information.

**Examples:**
- User names, emails
- Product prices
- Transaction records

👉 This is the **most valuable part** of the database.

---

## ✅ One-line Summary

> Database Software manages Databases → Databases contain Schemas → Schemas organize Tables → Tables store Data
