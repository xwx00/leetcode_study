# Introduction

`````python
class MyStack:
    def __init__(self):
        self.data = []  # store elements

    # Insert an element into the stack
    def push(self, x: int) -> None:
        self.data.append(x)

    # Checks whether the stack is empty or not
    def isEmpty(self) -> bool:
        return len(self.data) == 0

    # Get the top item from the stack
    def top(self) -> int:
        return self.data[-1]

    # Delete an element from the stack. Return True if successful
    def pop(self) -> bool:
        if self.isEmpty():
            return False
        self.data.pop()
        return True


if __name__ == "__main__":
    s = MyStack()
    s.push(1)
    s.push(2)
    s.push(3)

    for _ in range(4):
        if not s.isEmpty():
            print(s.top())
        print("true" if s.pop() else "false")
`````

## 150. Evaluate Reverse Polish Notation 

`````python
from typing import List

class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        stack = []

        for s in tokens:
            if s in '+-*/':
                b = stack.pop()
                a = stack.pop()

                if s == "+":
                    stack.append(a + b)
                elif s == "-":
                    stack.append(a - b)
                elif s == "*":
                    stack.append(a * b)
                elif s == "/":
                    stack.append(int(a/b)) 
            
            else:
                stack.append(int(s))

        return stack[0]
        
`````

# DFS

## Recursive Template

`````python
def dfs(cur, target, visited):
    # Base case
    if cur == target:
        return True
    
    # Traverse neighbors
    for nxt in cur.neighbors:   # assuming Node has attribute `neighbors`
        if nxt not in visited:
            visited.add(nxt)
            if dfs(nxt, target, visited):
                return True
    
    return False
`````

### Number of Islands

````python
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        m, n = len(grid), len(grid[0])
        nums = 0
        directions = [1, 0, -1, 0, 1]
        
                    
        def dfs(i,j):
            if grid[i][j] == "0":
                return 
            
            grid[i][j] = "0"
            
            for idx in range(4):
                nxt_i = i + directions[idx]
                nxt_j = j + directions[idx+1]
                
                if 0 <= nxt_i < m and 0 <= nxt_j < n and grid[nxt_i][nxt_j] == "1":
                    dfs(nxt_i,nxt_j)
                    
            return 
        
        for i in range(m):
            for j in range(n):
                if grid[i][j] == "1":
                    nums += 1
                    dfs(i,j)
        
        return nums
```` 
